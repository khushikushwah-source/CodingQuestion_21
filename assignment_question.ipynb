**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
def fact(n):
    if n==0:
        return 1
    else:
        return n * fact(n-1)

Num = int(input("Enter a number"))
print(f"factorial of {Num} is {fact(Num)}")
```
